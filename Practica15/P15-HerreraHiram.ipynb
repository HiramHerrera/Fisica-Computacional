```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def relajacion(phi,rho,V,tol = 1e-3):
    Nxdim,Nydim = np.shape(phi)
    phi_old = np.zeros(np.shape(phi))
    n = 0
    while (abs(phi_old-phi)>tol).any():
        phi_old = phi.copy()
        for i in range(1,Nxdim-1):
            for j in range(1,Nydim-1):
                if phi[i,j] == V or Phi[i,j] == -V:
                    continue
                else:
                    phi[i,j] = 0.25*(phi[i+1,j]+phi[i-1,j]+phi[i,j+1]+phi[i,j-1])-np.pi*rho[i,j]
        n+=1
    print('Iteraciones:',n) 
    return phi
```

```python
# Problema 1
V = 100.
Phi = np.zeros([100,100])
rho = np.zeros([100,100])
Phi[:,0] = V
Xdim,Ydim = np.shape(Phi)
X,Y = np.arange(Xdim),np.arange(Ydim)
plt.pcolor(X,Y,Phi,cmap='jet')
plt.colorbar()
plt.title('Condici√≥n Inicial')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
Phi = relajacion(Phi,rho,V)
```

```python
plt.pcolor(X,Y,Phi,cmap='jet')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
x,y = np.meshgrid(X,Y)
ax = plt.axes(projection='3d')
ax.plot_surface(x,y,Phi)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel(r'$\Phi(x,y)$')
plt.show()
```
