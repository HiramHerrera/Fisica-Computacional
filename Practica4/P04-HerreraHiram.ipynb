# 1. Partícula clásica en una caja unidimensional

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Accel(x):
    a = 0
    return a
def Euler(x0,v0,dt):
    v = v0 + Accel(x0)*dt
    x = x0 + v0 * dt
    return x,v
```

```python
#Condiciones Iniciales
t0 = 0
tf = 20
dt = 0.001
v0 = 2.
x0 = 0.
L = 10.
```

```python
 def Problema1(L,x0,v0,t0,tf,dt):
    x = x0
    v = v0
    V = []
    X = []
    T = []
    while t0<tf:
        x,v = Euler(x,v,dt)
        if x>=L or x<=0:
            v = -v
        t0 += dt
        V.append(v), X.append(x),T.append(t0)
    plt.plot(T,X)
    plt.xlabel('t (s)')
    plt.ylabel('x (m)')
    plt.show()
    plt.plot(T,V)
    plt.xlabel('t (s)')
    plt.ylabel('v (m/s)')
    plt.show()
    return 
```

```python
Problema1(L,x0,v0,t0,tf,dt)
```

```python
t0 = 0
tf = 30
dt = 0.005
v0 = 5.
x0 = 1.
L = 20.
Problema1(L,x0,v0,t0,tf,dt)
```

# 2. Partícula en una caja Bidimensional

```python
 def Problema2(Lx,Ly,x0,y0,vx0,vy0,t0,tf,dt):
    contadorx,contadory = 0,0
    x,y = x0,y0
    vx,vy = vx0,vy0
    Vx,Vy = [],[]
    X,Y = [],[]
    T = []
    while t0<tf:
        x,vx = Euler(x,vx,dt)
        y,vy = Euler(y,vy,dt)
        if x>=Lx or x<=0:
            vx = -vx
            contadorx +=1
        if y>=Ly or y<=0:
            vy = -vy
            contadory +=1
        t0 += dt
        Vx.append(vx), Vy.append(vy)
        X.append(x), Y.append(y),T.append(t0)
    print('Colisiones X:',contadorx)
    print('Colisiones Y:',contadory)
    plt.plot(T,X)
    plt.xlabel('t (s)')
    plt.ylabel('x (m)')
    plt.show()
    plt.plot(T,Vx)
    plt.xlabel('t (s)')
    plt.ylabel(r'$v_x (m/s)$')
    plt.show()
    plt.plot(T,Y)
    plt.xlabel('t (s)')
    plt.ylabel('y (m)')
    plt.show()
    plt.plot(T,Vy)
    plt.xlabel('t (s)')
    plt.ylabel(r'$v_y (m/s)$')
    plt.show()
    plt.plot(X,Y)
    plt.xlabel('x (m)')
    plt.ylabel('y (m)')
    plt.show()
    return 
```

```python
t0 = 0
tf = 20
dt = 0.001
vx0 = 3.
x0 = 1.
vy0 = 2.
y0 = 3.
Lx = 10.
Ly = 5.
Problema2(Lx,Ly,x0,y0,vx0,vy0,t0,tf,dt)
```

# 3. Caja con Campo eléctrico

Caja Unidimensional

```python
def Accel(E0,q,x):
    a = -q*E0
    return a
def Euler(E0,q,x0,v0,dt):
    v = v0 + Accel(E0,q,x0)*dt
    x = x0 + v0 * dt
    return x,v
```

```python
 def Problema3a(E0,q,L,x0,v0,t0,tf,dt):
    x,v = x0,v0    
    V,X,T = [],[],[]
    while t0<tf:
        x,v = Euler(E0,q,x,v,dt)
        if x>=L or x<=0:
            v = -v
        t0 += dt
        V.append(v), X.append(x),T.append(t0)
    plt.plot(T,X)
    plt.xlabel('t (s)')
    plt.ylabel('x (m)')
    plt.show()
    plt.plot(T,V)
    plt.xlabel('t (s)')
    plt.ylabel('v (m/s)')
    plt.show()
    return 
```

```python
t0 = 0
tf = 20
dt = 0.001
v0 = 2.
x0 = 0.
E0 = 1.
q = 2.
L = 10.
Problema3a(E0,q,L,x0,v0,t0,tf,dt)
```

```python
t0 = 0
tf = 20
dt = 0.001
v0 = 2.
x0 = 0.
E0 = 0.5
q = 2.
L = 1.9
Problema3a(E0,q,L,x0,v0,t0,tf,dt)
```
