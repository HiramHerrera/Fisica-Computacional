```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w0 = 1.
m = 1.
```

```python
def Accel(r):
    a = -w0 **2.*r
    return a

def Leapfrog(r0,v0,dt):
    Vmed = v0 + Accel(r0)*dt/2.
    r = r0 + Vmed*dt
    v = Vmed + Accel(r)*dt/2.
    return r,v

def Euler(r0,v0,dt):
    v = v0 + Accel(r0)*dt
    r = r0 + v0 * dt
    return r,v

def EulerCromer(r0,v0,dt):
    v = v0 + Accel(r0)*dt
    r = r0 + v * dt
    return r,v

def PuntoMed(r0,v0,dt):
    v = v0 + Accel(r0)*dt
    r = r0 + 0.5*(v+v0)*dt
    return r,v

def Energy(r,v):
    #Energia por unidad de masa
    K = 0.5*v**2.
    V = 0.5*w0**2.*r**2.
    return K+V
```

```python
dt = 0.01
x0 = 0.
v0 = 1.
t0 = 0.
tf = 10.
E0 = Energy(x0,v0)
print(E0)
```

```python
#LEAPFROG
X,V,T=[x0],[v0],[t0]
Err_LF = [0.]
x,v,t=x0,v0,t0
while t<tf:
    x,v=Leapfrog(x,v,dt)
    E = Energy(x,v)
    Error = abs((E-E0)/E0)
    t +=  dt
    X.append(x),V.append(V),T.append(t)
    Err_LF.append(Error)
plt.plot(T,X)
plt.xlabel('T (s)')
plt.ylabel('X (m)')
plt.show()

plt.plot(T,Err_LF)
plt.xlabel('T (s)')
plt.ylabel('Error de EnergÃ­a')
plt.show()
```

```python
#EULER
X,V,T=[x0],[v0],[t0]
Err_EU = [0.]
x,v,t=x0,v0,t0
while t<tf:
    x,v=Euler(x,v,dt)
    E = Energy(x,v)
    Error = abs((E-E0)/E0)
    t +=  dt
    X.append(x),V.append(V),T.append(t)
    Err_EU.append(Error)
plt.plot(T,X)
plt.xlabel('T (s)')
plt.ylabel('X (m)')
plt.show()
plt.plot(T,Err_EU)
plt.xlabel('T (s)')
plt.ylabel('Error de EnergÃ­a')
plt.show()
```

```python
#EULER-CROMER
X,V,T=[x0],[v0],[t0]
Err_EC = [0.]
x,v,t=x0,v0,t0
while t<tf:
    x,v=EulerCromer(x,v,dt)
    E = Energy(x,v)
    Error = abs((E-E0)/E0)
    t +=  dt
    X.append(x),V.append(V),T.append(t)
    Err_EC.append(Error)
plt.plot(T,X)
plt.xlabel('T (s)')
plt.ylabel('X (m)')
plt.show()
plt.plot(T,Err_EC)
plt.xlabel('T (s)')
plt.ylabel('Error de EnergÃ­a')
plt.show()
```
