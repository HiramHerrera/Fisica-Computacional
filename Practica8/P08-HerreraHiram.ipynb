Problema 1.1.4

Desarrollamos en series de Taylor $f(x_n)$ alrededor de $x*$ a primer orden:

$f(x_n) = f(x*) + f'(x*)(x_n-x*)$

Restando $f(x*)$ obtenemos:

$f(x_n) -f(x*) = f'(x*)(x_n-x*)$
1
Si tomamos el hecho de que $f(x*)= x*$ y $f(x_{n}) = x_{n+1}$ entonces podemos escribir:

$x_{n+1}- x* = f'(x*)(x_n-x*)$ 

Aplicando valores absolutos y despejando $f'(x*)$ tenemos

$|f'(x*)| < \left| \frac{x_{n+1}- x*}{x_n-x*}\right|$

Así para que el método converja a $x*$ necesitamos que 

$|f'(x*)| < 1$

Problema 1.1.5

Si $f(x)$ es invertible y además $|f'(x*)| < 1$, podemos reescribir $f(x)=x$ como $x = f^{-1}(x)$, así igualando las expresiones obtenemos que $f(x)=f^{-1}(x)$, y por lo tanto  $f' (x)=f'^{-1} (x)$ por lo que se cumpliría que $|f'^{-1}(x*)|< 1$, garantizando que la inversa de la ecuación también converge.

```python
#Problema 1
def f(x):
    return 2.*x**2.-10.*x+3
```

```python
#1 a)
X = np.linspace(-6,4,1000)
plt.plot(X,f(X),'b')
plt.xlabel('x')
plt.axhline(y=0.0,color ='k')
plt.ylabel('f(x)')
plt.show()
```

```python
# 1 b)
a = -6
b = 4
tol = 1E-7
i = 1
FA = f(a)
maxsteps=100

while (i<maxsteps):
    p = a + (b-a)/2.0
    FP = f(p)
    Err = (b-a)/2.0
    if (FP == 0.0) or (Err < tol):
        break
    i +=1
    if (f(a)*f(p) > 0):	
        a = p 
        FA = FP
    else: 
        b = p
print(p,Err,i)
```

$\textbf{Problema 2.}$

Tenemos $I(\lambda) = \frac{2\pi hc^2 \lambda^{-5}}{e^{hc/ \lambda k_B T}-1}$, derivamos con respecto de $\lambda$ entonces e igualamos a cero, entonces:

$2\pi hc^2\left(\frac{-5\lambda^{-6}(e^{hc/ \lambda k_B T}-1)+ \frac{hc} {\lambda^2 k_B T}\lambda^{-5}(e^{hc/ \lambda k_B T})}{(e^{hc/ \lambda k_B T}-1)^2}\right) = 0$

De lo anterior 

$-5\lambda^{-6}(e^{hc/ \lambda k_B T}-1)+ \frac{hc} {k_B T}\lambda^{-7}(e^{hc/ \lambda k_B T})= 0 $


Distribuyendo y factorizando un $\lambda^{-6}$ obtenemos

$-5e^{hc/ \lambda k_B T}+5 + \frac{hc} {k_B T\lambda}e^{hc/ \lambda k_B T}= 0 $

Dividiendo sobre $e^{hc/ \lambda k_B T}$ y reacomodamos obtenemos

$5e^{-hc/ \lambda k_B T} + \frac{hc} {\lambda k_B T} - 5= 0$

Si  $x = hc/ \lambda k_B T$ la ecuación queda como:

$5e^{-x} + x - 5= 0$


```python
# 1)
def g(x):
    return 5.*np.exp(-x) + x - 5. 


a = 1
b = 5
tol = 1E-6
i = 1
FA = g(a)
maxsteps=100

while (i<maxsteps):
    p = a + (b-a)/2.0
    FP = g(p)
    Err = (b-a)/2.0
    if (FP == 0.0) or (Err < tol):
        break
    i +=1
    if (g(a)*g(p) > 0):	
        a = p 
        FA = FP
    else: 
        b = p
print(p,Err,i)
```

$\textbf{Problema 3.}$

Usando las unidades:

$[E]=1eV=1.60218\times 10^{-19}J$  - Energíá

$[T]=6.58210\times 10^{-16} s$ - Tiempo

$[L]=10^{-9}m$ - Longitud

$[M]=6.941099\times10^{-32}$ - Masa

Y haciendo las conversiones correspondientes obtenemos:

$\hbar=1$, $\omega=1$, $m=13.1235$ y $V=20$ 

```python
hbar = 1.
m =  13.1235
w =  1.
E = np.linspace(0.000001,19.999,1000)
V = 20.


Y1 = np.tan(np.sqrt(0.5*w**2.*m*E/hbar**2.))
Y2 = np.sqrt((V-E)/E)
Y3 = -np.sqrt(E/(V-E))


plt.plot(E,Y1,'k',label=r'$y_1$')
plt.plot(E,Y2,'b--',label=r'$y_2$')
plt.plot(E,Y3,'r--',label=r'$y_3$')
plt.xlabel('E (eV)')
plt.axhline(y=0,color='k')
plt.ylabel('f(E)')
plt.ylim(-5,10)
plt.legend()
plt.show()
```

De la figura vemos que el sexto nivel de energía $E_5$ al rededor de 11 eV

```python
def F(En,n):
    if n%2==0:
        Fun = np.tan(np.sqrt(0.5*w**2*m*En/hbar**2.))-np.sqrt((V-En)/En)
    else:
        Fun = np.tan(np.sqrt(0.5*w**2*m*En/hbar**2.))+np.sqrt(En/(V-En))
    return Fun


intervals = [0.00001,0.5,2.5,4.5,6.,10.,13.,17.5,20.]
tol = 1E-4
n = 0
maxsteps=100
i=0
for j in range(8):
    a = intervals[j]
    b = intervals[j+1]
    FA = F(a,n)
    while (i<maxsteps):
        p = a + (b-a)/2.0
        FP = F(p,n)
        Err = (b-a)/2.0
        if (FP == 0.0) or (Err < tol):
            break
        i +=1
        if (F(a,n)*F(p,n) > 0):
            a = p 
            FA = FP
        else: 
            b = p
    print('n:',n,'Energía:',p,'eV')
    n+=1
```
