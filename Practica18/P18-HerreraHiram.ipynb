# Random Walk

```python
def randomwalk1D(a,b,limit,steps):
    x = 0.
    X = [x]
    T = np.arange(steps)
    for i in range(steps-1):
        dx = (b-a)*random()+a
        x += dx
        if x >= limit:
            x = 0
        if x <= 0:
            x = abs(x)
        X.append(x)
    plt.plot(T,X)
    plt.plot(T,X,'ko',markersize=3)
    plt.xlabel('Pasos')
    plt.ylabel('x')
    #plt.savefig('Figures/Walk1D_%2.1f_%2.1f_%i.png'%(a,b,steps))
    plt.show()    
```

```python
randomwalk1D(0,1,50,32)
randomwalk1D(0,1,50,256)
randomwalk1D(-1,1,50,32)
randomwalk1D(-1,1,50,256)
```

```python
def randomwalk2D(a,b,limit,steps):
    x,y = 0.,0.
    X,Y = [x],[y]
    T = np.arange(steps)
    for i in range(steps-1):
        dx,dy = (b-a)*random()+a,(b-a)*random()+a
        x += dx
        y += dy
        if x >= limit:
            x = limit-dx
        if x <= 0:
            x = abs(x)
        if y >= limit:
            y = limit-dy
        if y <= 0:
            y = abs(y)
        X.append(x),Y.append(y)
    plt.plot(X,Y)
    plt.plot(X,Y,'ko',markersize=3)
    plt.xlabel('x')
    plt.ylabel('y')
    #plt.savefig('Figures/Walk2D_%2.1f_%2.1f_%i.png'%(a,b,steps))
    plt.show()   
```

```python
randomwalk2D(0,1,10,32)
randomwalk2D(0,1,10,256)
randomwalk2D(-.1,.1,10,32)
randomwalk2D(-.1,.1,10,256)
```

```python
def randomwalk3D(a,b,limit,steps):
    x,y,z = 0.,0.,0.
    X,Y,Z = [x],[y],[z]
    T = np.arange(steps)
    for i in range(steps-1):
        dx,dy,dz = (b-a)*random()+a,(b-a)*random()+a,(b-a)*random()+a
        x += dx
        y += dy
        z += dz
        if x >= limit:
            x = limit-dx
        if x <= 0:
            x = abs(x)
        if y >= limit:
            y = limit-dy
        if y <= 0:
            y = abs(y)
        if z >= limit:
            z = limit-dz
        if z <= 0:
            z = abs(z)

        X.append(x),Y.append(y),Z.append(z)
    ax = plt.axes(projection='3d')
    ax.plot3D(X,Y,Z)
    ax.plot3D(X,Y,Z,'ko',markersize=3)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    #plt.savefig('Figures/Walk3D_%2.1f_%2.1f_%i.png'%(a,b,steps))
    plt.show() 
```

```python
randomwalk3D(-2,2,30,32)
randomwalk3D(-2,2,30,256)
```

# Variables Aleatorias

```python
def Accel(r):
    a = 20.*random()-10.
    return a

def Euler(r0,v0,dt):
    v = v0 + Accel(r0)*dt
    r = r0 + v0 * dt
    return r,v

def Leapfrog(r0,v0,dt):
    Vmed = v0 + Accel(r0)*dt/2.
    r = r0 + Vmed*dt
    v = Vmed + Accel(r)*dt/2.
    return r,v
```

```python
dt = 0.01
steps = 300
v = 1.
x = 0.
t = 0.

Xarr,T = [x],[t]
for i in range(steps+1):
    x,v = Euler(x,v,dt)
    t+=dt
    Xarr.append(x),T.append(t)

plt.plot(T,Xarr,'b',label='Euler')
    
v = 1.
x = 0.
t = 0.

Xarr,T = [x],[t]
for i in range(steps+1):
    x,v = Leapfrog(x,v,dt)
    t+=dt
    Xarr.append(x),T.append(t)

plt.plot(T,Xarr,'k',label='Leap-Frog')
plt.xlabel('t (s)')
plt.ylabel('x')
plt.legend()
#plt.savefig('Figures/RandomUni.png')
plt.show()
```
