```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def RK5(f,x,t,h):
    k1 = h*f(x,t)
    k2 = h*f(x+k1/4.,t+h/4.)
    k3 = h*f(x+k1/8.+k2/8.,t+h/4.)
    k4 = h*f(x-k2/2.+k3,t+h/2.)
    k5 = h*f(x+3.*k1/16 + 9*k4/16,t+3.*h/4.)
    k6 = h*f(x-3.*k1/7+2.*k2/7.+12.*k3/7.-12.*k4/7.+8.*k5/7.,t+h)
    x = x + (7.*k1+32.*k3+12.*k4+32.*k5+7.*k6)/90.
    return x

def f(r,t,sigma = 10., betha = 8./3.):
    x,y,z = r[0], r[1],r[2]
    fxd = sigma * (y-x)
    fyd = x*(rho-z)-y
    fzd = x*y-betha*z
    return np.array([fxd,fyd,fzd])

def Lorenz(x0,y0,z0,t0,tf,N=10000):
    h = (tf - t0)/N
    X,Y,Z,T= [x0],[y0],[z0],[t0]
    r = np.array([x0,y0,z0])
    for i in range(1,N+1):
        r = RK5(f,r,t0,h)
        t0 += h
        X.append(r[0]),Y.append(r[1]),Z.append(r[2]),T.append(t0)
    return X,Y,Z,T
```

```python
#4.1
rho = 15.
X,Y,Z,T = Lorenz(10.,10.,22.,0.,13.)
plt.plot(T,X)
plt.xlabel('t')
plt.ylabel('x')
plt.show()
```
